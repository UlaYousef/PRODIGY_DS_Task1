```python
sns.set(palette="rainbow")
f, axes = plt.subplots(3, 2,figsize=(8,12))
sns.histplot(df['Weight'],ax = axes[0,0],kde=True)
sns.histplot(df['Length1'],ax=axes[0,1],kde=True)
sns.histplot(df['Length2'],ax=axes[1,0],kde=True)
sns.histplot(df['Length3'],ax=axes[1,1],kde=True)
sns.histplot(df['Height'],ax=axes[2,0],kde=True)
sns.histplot(df['Width'],ax=axes[2,1],kde=True)
plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10,10))
andrews_curves(df,"Species")
plt.show()
```

```python
sns.pairplot(df,hue='Species',palette='inferno')
```

```python
sns.set(palette='viridis')
f,axes=plt.subplots(3,2,figsize=(25,10))
sns.boxplot(x=df['Species'],y=df['Weight'],ax=axes[0,0])
sns.boxplot(x=df['Species'],y=df['Length1'],ax=axes[0,1])
sns.boxplot(x=df['Species'],y=df['Length2'],ax=axes[1,0])
sns.boxplot(x=df['Species'],y=df['Length3'],ax=axes[1,1])
sns.boxplot(x=df['Species'],y=df['Height'],ax=axes[2,0])
sns.boxplot(x=df['Species'],y=df['Width'],ax=axes[2,1])
plt.show()
```

```python
sns.set(palette='viridis')
f, axes = plt.subplots(3, 2,figsize=(20,8))
sns.violinplot(data=df, y="Weight",x='Species',ax= axes[0,0])
sns.violinplot(data=df, y="Length1",x='Species',ax= axes[0,1])
sns.violinplot(data=df, y="Length2",x='Species',ax= axes[1,0])
sns.violinplot(data=df, y="Length3",x='Species',ax= axes[1,1])
sns.violinplot(data=df, y="Height",x='Species',ax= axes[2,0])
sns.violinplot(data=df, y="Width",x='Species',ax= axes[2,1])
plt.show()
```
